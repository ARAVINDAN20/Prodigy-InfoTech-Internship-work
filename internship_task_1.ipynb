```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
# url = 'https://api.worldbank.org/v2/en/indicator/SP.POP.TOTL?downloadformat=csv'
# df = pd.read_csv(url, skiprows=4)
df = pd.read_csv("Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_363248.csv")

print(df.head())
df.dropna()
df = df.iloc[:, :-1]
```

```python
# prompt: set the columns name is Country Code	Region	IncomeGroup	SpecialNotes	TableName for each columns for the above data set and display

df.columns = ["Country Code", "Region", "IncomeGroup", "SpecialNotes", "TableName"]
df
```

```python
plt.figure(figsize=(14, 8))
plt.bar(df['Country Code'][:20], df["TableName"][:20])
plt.xlabel('Country')
plt.ylabel('Total Population')
plt.title(f'Top 20 Countries by Population in 2020')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
